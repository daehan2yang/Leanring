```python
# MLP model: # input (28x28), 2 hidden layers (128, 64), output (10)
# Dataset 가져와서 0~255 픽셀값을 0~1 사이의 값으로 정규화
# training data는 60,000개, test data는 10,000개

import torch
from torchvision import datasets, transforms
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, ConfusionMatrixDisplay
import pandas as pd
import matplotlib.pyplot as plt

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# 이미지를 텐서로 변환
transform = transforms.ToTensor()

# 학습용 데이터셋
train_dataset = datasets.MNIST(
    root='./data',          # 다운로드 및 저장
    train=True,             # 학습 데이터
    download=True,          # 없으면 자동 다운로드
    transform=transform     # transform 전처리 적용
)

# 테스트용 데이터셋
test_dataset = datasets.MNIST(root='./data', train=False, download=True, transform=transform)

# 텐서로 꺼내기 (DataLoader 없이 전체 텐서 사용)
X_train = train_dataset.data.view(-1, 28*28).float() / 255.0
y_train = train_dataset.targets

X_test = test_dataset.data.view(-1, 28*28).float() / 255.0
y_test = test_dataset.targets
```

```python
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
# 초기화 함수들
init_methods = {
    'xavier': lambda m: nn.init.xavier_uniform_(m.weight),
    'he': lambda m: nn.init.kaiming_uniform_(m.weight, nonlinearity='relu'),
    'normal': lambda m: nn.init.normal_(m.weight, mean=0.0, std=0.01),
    'uniform': lambda m: nn.init.uniform_(m.weight, a=-0.1, b=0.1)
}

# 모델 생성 함수
def get_model(init_name):
    model = nn.Sequential(
        nn.Linear(784, 128),
        nn.ReLU(),
        nn.Linear(128, 64),
        nn.ReLU(),
        nn.Linear(64, 10)
    )
    for m in model:
        if isinstance(m, nn.Linear):
            init_methods[init_name](m)
    return model.to(device)

# 학습 함수
def train(model, X, y, criterion, optimizer, epochs):
    for _ in range(epochs):
        model.train()
        optimizer.zero_grad()
        output = model(X)
        loss = criterion(output, y)
        loss.backward()
        optimizer.step()

# 평가 함수
def evaluate(model, X, y):
    model.eval()
    with torch.no_grad():
        output = model(X)
        pred = output.argmax(dim=1)
        acc = accuracy_score(y.cpu(), pred.cpu())
        prec = precision_score(y.cpu(), pred.cpu(), average='weighted')
        rec = recall_score(y.cpu(), pred.cpu(), average='weighted')
        f1 = f1_score(y.cpu(), pred.cpu(), average='weighted')
        return acc, prec, rec, f1

# 실험 환경
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
X_train_dev, y_train_dev = X_train.to(device), y_train.to(device)
X_test_dev, y_test_dev = X_test.to(device), y_test.to(device)

# 하이퍼파라미터 조합
learning_rates = [0.001, 0.005, 0.01]
epoch_list = [5, 10, 15]

# 전체 실험
results = []
for init_name in init_methods:
    for lr in learning_rates:
        for epoch in epoch_list:
            model = get_model(init_name)
            criterion = nn.CrossEntropyLoss()
            optimizer = optim.Adam(model.parameters(), lr=lr)
            train(model, X_train_dev, y_train_dev, criterion, optimizer, epoch)
            acc, prec, rec, f1 = evaluate(model, X_test_dev, y_test_dev)
            print(f"초기화: {init_name}, LR: {lr}, Epochs: {epoch}")
            print(f"  Accuracy: {acc:.4f}, | Precision: {prec:.4f}, | Recall: {rec:.4f}, | F1: {f1:.4f}")
            results.append({
                "Init": init_name,
                "LR": lr,
                "Epochs": epoch,
                "Accuracy": acc,
                "Precision": prec,
                "Recall": rec,
                "F1 Score": f1
            })

# 결과 시각화
results_df = pd.DataFrame(results)
plt.figure(figsize=(10,6))
for init_name in init_methods:
    subset = results_df[results_df['Init'] == init_name]
    plt.plot(subset['Epochs'], subset['Accuracy'], marker='o', label=init_name)
plt.title("Epoch Average")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# 결과 분석
df_results = pd.DataFrame(results)
best = results_df.sort_values(by='Accuracy', ascending=False).iloc[0]
worst = results_df.sort_values(by='Accuracy').iloc[0]
avg_row = results_df.iloc[(results_df['Accuracy'].mean() - results_df['Accuracy']).abs().argsort().iloc[0]]

print("\n최적의 파라미터 조합:")
print(f"Init = {best['Init']}, Learning Rate = {best['LR']}, Epochs = {int(best['Epochs'])}")
print(f"정확도 (Accuracy): {best['Accuracy']:.8f}")
print(f"정밀도 (Precision): {best['Precision']:.8f}")
print(f"재현율 (Recall): {best['Recall']:.8f}")
print(f"F1 점수: {best['F1 Score']:.8f}")

print("\n평균 근접 조합:")
print(f"Init = {avg_row['Init']}, Learning Rate = {avg_row['LR']}, Epochs = {int(avg_row['Epochs'])}")
print(f"정확도 (Accuracy): {avg_row['Accuracy']:.8f}")
print(f"정밀도 (Precision): {avg_row['Precision']:.8f}")
print(f"재현율 (Recall): {avg_row['Recall']:.8f}")
print(f"F1 점수: {avg_row['F1 Score']:.8f}")

print("\n최악의 조합:")
print(f"Init = {worst['Init']}, Learning Rate = {worst['LR']}, Epochs = {int(worst['Epochs'])}")
print(f"정확도 (Accuracy): {worst['Accuracy']:.8f}")
print(f"정밀도 (Precision): {worst['Precision']:.8f}")
print(f"재현율 (Recall): {worst['Recall']:.8f}")
print(f"F1 점수: {worst['F1 Score']:.8f}")

# 성능 비교 시각화 (막대 + 선 혼합)
best_acc = best['Accuracy']
worst_acc = worst['Accuracy']
avg_acc = avg_row['Accuracy']

plt.figure(figsize=(8, 5))
plt.bar(['Best', 'Worst'], [best_acc, worst_acc], color=['green', 'red'])
plt.plot(['Best', 'Worst'], [avg_acc, avg_acc], marker='o', color='blue', label='Average Trend')
plt.title("Accuracy Comparison: Best vs Avg vs Worst Hyperparams")
plt.ylabel("Accuracy")
plt.ylim(0, 1.0)
plt.grid(True, axis='y')
plt.legend()
plt.show()
```

```python
# 최적 모델로 혼동 행렬 출력
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

model = get_model(best['Init'])
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=best['LR'])
train(model, X_train_dev, y_train_dev, criterion, optimizer, int(best['Epochs']))

model.eval()
with torch.no_grad():
    output = model(X_test_dev)
    pred_value = output.argmax(dim=1)
    true_value = y_test_dev

cm = confusion_matrix(true_value.cpu(), pred_value.cpu())
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[str(i) for i in range(10)])
plt.figure(figsize=(8, 6))
disp.plot(cmap='Blues', values_format='d')
plt.title(" Confusion Matrix (Best Hyperparameters)")
plt.grid(False)
plt.show()
```

```python
import numpy as np
from sklearn.metrics import precision_score, recall_score, f1_score
import matplotlib.pyplot as plt

ratios = [0.1, 0.3, 0.5]
subset_results = []

for ratio in ratios:
    subset_size = int(len(X_train) * ratio)
    indices = torch.randperm(len(X_train))[:subset_size]
    X_sub = X_train[indices].to(device)
    y_sub = y_train[indices].to(device)

    model = get_model(best['Init'])
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=float(best['LR']))
    print(f"\n Training with {int(ratio*100)}% of data ({subset_size} samples)")
    train(model, X_sub, y_sub, criterion, optimizer, int(best['Epochs']))

    model.eval()
    with torch.no_grad():
        output = model(X_test_dev)
        preds = output.argmax(dim=1)

        acc = accuracy_score(y_test_dev.cpu(), preds.cpu())
        prec = precision_score(y_test_dev.cpu(), preds.cpu(), average='weighted')
        rec = recall_score(y_test_dev.cpu(), preds.cpu(), average='weighted')
        f1 = f1_score(y_test_dev.cpu(), preds.cpu(), average='weighted')

        print(f"     Accuracy: {acc:.4f}, | Precision: {prec:.4f}, | Recall: {rec:.4f}, | F1: {f1:.4f}")

        subset_results.append({
            "Ratio": int(ratio * 100),
            "Accuracy": acc,
            "Precision": prec,
            "Recall": rec,
            "F1 Score": f1
        })

```

```python
# 결과 출력
results_df = pd.DataFrame(subset_results)
print("\n 학습 비율별 성능 비교:")
print(results_df)

# 성능 시각화 (최종 결과에서만 출력)
plt.figure()
plt.plot(results_df['Ratio'], results_df['Accuracy'], marker='o', label='Accuracy')
plt.plot(results_df['Ratio'], results_df['Precision'], marker='o', label='Precision')
plt.plot(results_df['Ratio'], results_df['Recall'], marker='o', label='Recall')
plt.plot(results_df['Ratio'], results_df['F1 Score'], marker='o', label='F1 Score')
plt.title('Performance by Training Data Ratio')
plt.xlabel('Training Data Ratio (%)')
plt.ylabel('Score')
plt.xticks(results_df['Ratio'])
plt.legend()
plt.grid(True)
plt.show()
```
